```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "../Instructions/Pymaceuticals/data/mouse_drug_data.csv"
clinical_trial_data_to_load = "../Instructions/Pymaceuticals/data/clinicaltrial_data.csv"

mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)
# join at Mouse ID

mouse_clinical_merge_data = pd.merge(mouse_drug_data, clinical_trial_data, on='Mouse ID')
# mouse_clinical_merge_data.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol_groupby = mouse_clinical_merge_data.groupby(["Drug","Timepoint"]).mean().drop(columns='Metastatic Sites')
mean_tumor_vol_df = pd.DataFrame(mean_tumor_vol_groupby)
mean_tumor_vol_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
mean_tumor_stderr = mouse_clinical_merge_data.groupby(["Drug","Timepoint"]).sem().drop(columns=["Metastatic Sites", "Mouse ID"])
# mean_tumor_stderr.head()
mean_tumor_stderr_df = pd.DataFrame(mean_tumor_stderr)
mean_tumor_stderr_df.head()

# Convert to DataFrame

# Preview DataFrame

```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
# pivot the table to get Drug as Columns and Timepoint to be the Index
mean_tumor_vol_df.reset_index()

tumor_pivot_mean = pd.pivot_table(mean_tumor_vol_df, index=['Timepoint'], columns=['Drug'], values='Tumor Volume (mm3)')
tumor_pivot_mean.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Capomulin"], yerr=tumor_pivot_mean["Capomulin"], color='r', marker='o', markersize=5, linestyle='dashed', linewidth=0.40)
plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Infubinol"], yerr=tumor_pivot_mean["Infubinol"], color='b', marker='^', markersize=5, linestyle='dashed', linewidth=0.40)
plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Ketapril"], yerr=tumor_pivot_mean["Ketapril"], color='g', marker='s', markersize=5, linestyle='dashed', linewidth=0.40)
plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Placebo"], yerr=tumor_pivot_mean["Placebo"], color='k', marker='d', markersize=5, linestyle='dashed', linewidth=0.40)
# new_stderr_df.reset_index()

# plot needs things
plt.title("Tumor Response to Treatment vs Placebo")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc='best', fontsize='small', fancybox=True)


# Save the Figure
plt.savefig("Fig1.png")

plt.show
```

```python
survival_count = mouse_clinical_merge_data.groupby(["Drug", "Timepoint"]).count()["Tumor Volume (mm3)"]

# convert to DF and add survival count
survival_count = pd.DataFrame({"Mouse Count": survival_count})
survival_count.head().reset_index()
```

```python
survival_count = survival_count.reset_index()
pivot_survival_count = survival_count.pivot(index="Timepoint", columns="Drug")["Mouse Count"]

pivot_survival_count.head()
```

```python
# find the survival rate (%) for each treatment
plt.plot(100 * pivot_survival_count["Capomulin"] / 25, "ro", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * pivot_survival_count["Infubinol"] / 25, "b^", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * pivot_survival_count["Ketapril"] / 25, "gs", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * pivot_survival_count["Placebo"] / 25 , "kd", linestyle="dashed", markersize=6, linewidth=0.50)

plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="small", fancybox=True)

# save figure
plt.savefig("Fig3.png")

plt.show()
```
